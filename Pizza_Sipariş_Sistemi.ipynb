```python
# Pizza Sipariş Sistemi

# Gerekli kütüphanelerin yüklenmesi.

from datetime import datetime #  Sipariş zamanının anlık olarak alınması maksadıyla.

import csv #  Sipariş bilgilerinin csv uzantılı oluşturulacak dosyaya yazılabilmesi / okunabilmesi maksadıyla. 

from tabulate import tabulate #  Çıktının daha iyi görüntülenebilmesi maksadıyla.  
```

```python
# Pizza isimli üst sınıfın oluşturulması ve bu sınıfın içindeki encapsulation(kapsülleme) 
# için get_description() ve get_cost() metodların tanımlanamsı.

class Pizza:
    
    def get_description(self):
        return self.__class__.name
    
    def get_cost(self):
        return self.__class__.cost

# Klasik, Margarita, Türk Pizza ve Sade Pizza çeşitlerine ait alt sınıfların oluşturulması.
    
class Klasik(Pizza):
    name = "Klasik Pizza"
    cost = 120.0
    
class Margarita(Pizza):
    name = "Margarita Pizza"
    cost = 160.0
    
class Türk(Pizza):
    name = "Türk Pizza"
    cost = 150.0
    
class Sade(Pizza):
    name = "Sade Pizza"
    cost = 90.0
    
# Decorator sınıfının oluşturulması.

class Decorator(Pizza):
    
    def __init__(self, pizza):
        self.component = pizza
        
    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)
    
    def get_description(self):
        return self.component.get_description() + " " + Pizza.get_description(self)
        
# Zeytin, mantar, keçi peyniri, et, soğan ve mısır soslarına ait alt sınıfların oluşturulması.  
        
class Zeytin(Decorator):
    name = "(Zeytin Soslu)"
    cost = 10.0    
    def __init__(self, pizza):
        Decorator.__init__(self, pizza)

class Mantar(Decorator):
    name = "(Mantar Soslu)"
    cost = 10.0
    def __init__(self, pizza):
        Decorator.__init__(self, pizza)

class Keçi_Peyniri(Decorator):
    name = "(Keçi Peyniri Soslu)"
    cost = 20.0
    def __init__(self, pizza):
        Decorator.__init__(self, pizza)
        
class Et(Decorator):
    name = "(Et Soslu)"
    cost = 20.0
    def __init__(self, pizza):
        Decorator.__init__(self, pizza)

class Soğan(Decorator):
    name = "(Soğan Soslu)"    
    cost = 5.0
    def __init__(self, pizza):
        Decorator.__init__(self, pizza)
        
class Mısır(Decorator):
    name = "(Mısır Soslu)"    
    cost = 5.0
    def __init__(self, pizza):
        Decorator.__init__(self, pizza)
   
```

```python
# sipariş zamanının hesaplanarak değişkene atanmasına ilişkin fonksiyon.

def zaman_bilgisi():
    sipariş_zamanı = datetime.strftime(datetime.now(),"%c")
    return sipariş_zamanı
```

```python
# Orders_Database.csv isimli dosyanın ve dosyadaki sütun başlıklarını oluşturan fonksiyon.

def dosya_oluştur():

    başlıklar = ["Ad ve Soyad", "TC Kimlik No", "Kredi Kart No", "Kredi Kart Şifre", \
                 "Sipariş Zamanı", "Pizza Çeşiti ve Fiyat Bilgisi"]
    with open('Orders_Database.csv', mode='a') as Orders:
            yazici = csv.writer(Orders)
            yazici.writerow(başlıklar)
```

```python
# Orders_Database.csv isimli dosya içeriğini okuyan fonksiyon.

def dosya_oku():
    
    satirlar=[]
    with open("Orders_Database.csv") as dosyam:
      okuyucu = csv.reader(dosyam)
      for satir in okuyucu:
        satirlar.append(satir)
    print(tabulate(satirlar))
```
